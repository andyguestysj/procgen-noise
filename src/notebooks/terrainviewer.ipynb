```python
from ipywidgets import interact, IntSlider, FloatSlider
from project.core.game import Game
```

```python
def interactive_terrain():
    return interact(
        Game.plot_terrain,
        size=IntSlider(value=33, min=1, max=129, step=16, description="Size"),
        seed=IntSlider(value=11, min=0, max=100, step=1, description="Seed"),
        disp=FloatSlider(value=1.0, min=0.1, max=2.5, step=0.05, description="Displacement"),
        roughness=FloatSlider(value=0.55, min=0.3, max=0.7, step=0.01, description="Roughness"),
    )
```

```python
interactive_terrain()
```
